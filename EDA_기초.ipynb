```python
# 필요한 라이브러리 임포트
import pandas as pd # 데이터 분석
import numpy as np # 숫자 분석
from sklearn.datasets import load_iris
```

```python
# iris 데이터셋 불러오기
iris = load_iris()
```

```python
iris
```

#pandas dataFrame

```python
#데이터프레임으로 변환
df = pd.DataFrame(iris.data,columns=['sepal_length','sepal_width','petal_length','petal_width'])
df['species']= pd.Categorical.from_codes(iris.target,iris.target_names)
```

```python
#데이터 확인
print("=== 데이터 상위 5행 ===")
print(df.head())
```

```python
print('\n === 데이터 형태 (행,열) ===')
print(df.shape)
```

```python
print('\n === 컬럼 유형 확인 ===')
print(df.dtypes)
```

```python
print('\n === 결측치 확인 ===')
print(df.isnull().sum())
```

```python
print('/n === 종류별 데이터 개수 === ')
print(df['species'].value_counts())
```

#평균 (Mean)


```python
#단일 컬럼 평균
print('꽃받침 길이 평균: ',df['sepal_length'].mean())
#print(f"꽃받침 길이 평균: {df['sepal_length'].mean():.2f}")
```

```python
#전체 숫자형 컬럼 평균
print('\n전체 컬럼 평균')
print(df.mean(numeric_only= True))#numeric_only= True 숫자형 칼럼(정수,실수)만 계산
```

```python
#종류별 평균 비교
print('\n 종류별 꽃잎 길이 평균')
print(df.groupby('species')['petal_length'].mean())
```

```python
#종류별 평균 비교
print('\n 종류별 꽃잎 길이 평균')
print(df.groupby('species')['petal_length'].mean())
```

#중앙값(Median)

```python
#단일 칼럼 중앙값
print('꽃받침 길이 중앙값:',df['sepal_length'].median())
```

```python
#전체 컬럼 중앙값
print("\n 전체 칼럼 중앙값:")
print(df.median(numeric_only = True))
```

```python
#평균 vs 중앙값 비교
print('\n === 평균 vs 중앙값 ===')
comparison = pd.DataFrame({
    '평균': df.mean(numeric_only = True).round(2),
    '중앙값': df.median(numeric_only = True).round(2)
})
print(comparison)

```

### 넘파이(NumPY)의 기술통계 함수

```python
import numpy as np

data = df['sepal_length'].values #넘파이 배열로 변환
print(data)
```

```python
print('===NumPy 기술 통계 함수 ===')
print(f'np.mean()평균:{np.mean(data):.4f}')
print(f'np.median()중앙값:{np.median(data):.4f}')
print(f'np.min()최솟값:{np.min(data):.4f}')
print(f'np.max()최댓값:{np.max(data):.4f}')
print(f'np.var()분산(모집단):{np.var(data):.4f}') #모분산(n으로 나눔)
print(f'np.var(ddof=1) 분산(표본):{np.var(data,ddof=1):.4f}')#표본분산(n-1로 나눔)
print(f'np.std()표준편차(모집단):{np.std(data):.4f}')
print(f'np.std(ddof=1)표준편차(표본):{np.std(data,ddof=1):.4f}')
print(f'np.percentile(25)Q1 1사분위:{np.percentile(data,25):.4f}')
print(f'np.percentile(50)Q2 2사분위:{np.percentile(data,50):.4f}')
print(f'np.percentile(75)Q3 3사분위:{np.percentile(data,75):.4f}')
print(f'np.ptp()범위(max-min):{np.ptp(data):.4f}')
```

```python
# 다중 백분위수 한번에 계산
print('\n === 다중 백분위수 ===')
percentiles = np.percentile(data,[10,25,50,75,90])# 10%,25%,50% ...
percentiles
```

```python
for p,v in zip([10,25,50,75,90],percentiles):
  print(f'{p}번째 백분위수:{v:.2f}')
```
