* **ì£¼ìš” ì—°ì‚°ì**:

````python
a,b,c= 10,20,20
print(f'{a}=={b}:{a==b}')
print(f'{a}!={b}:{a!=b}')
print(f'{a}>{b}:{a>b}')
print(f'{b}<{c}:{b<c}')
print(f'{a}>={b}:{a>=b}')
print(f'{b}<={c}:{b<=c}')
````

## 5. [ì‹¤ìŠµ ì˜ˆì œ]

### ğŸŒ¡ï¸ ì˜¨ë„ ë³€í™˜ê¸° (ì˜¨ë„ ì¢…ë¥˜ í†µí•©í˜•)
1. **ì…ë ¥**: ì‚¬ìš©ìë¡œë¶€í„° ì˜¨ë„ ì¢…ë¥˜(ì„­ì”¨ ë˜ëŠ” í™”ì”¨)ì™€ ì˜¨ë„ ê°’ì„ ì…ë ¥ë°›ìŠµë‹ˆë‹¤.
2. **íŒë‹¨ ë° ê³„ì‚°**:
    * `if kind == 'ì„­ì”¨':` ì´ë©´ í™”ì”¨ ê³µì‹(`temp * 1.8 + 32`)ìœ¼ë¡œ ê³„ì‚°í•©ë‹ˆë‹¤.

````python
kind = input('ì˜¨ë„ ì¢…ë¥˜(ì„­ì”¨ ë˜ëŠ”  í™”ì”¨): ')

if kind == 'ì„­ì”¨':
  celsius = int(input('ì„­ì”¨ì˜¨ë„:'))
  fahrenheit = celsius * 1.8 + 32
  print(f'í™”ì”¨ì˜¨ë„ë¡œ ë³€í™˜í•œ ê°’:{fahrenheit}')
else:
  fahrenheit = int(input('í™”ì”¨ì˜¨ë„:'))
  celsius = (fahrenheit - 32) / 1.8
  print(f'ì„­ì”¨ì˜¨ë„ë¡œ ë³€í™˜í•œ ê°’:{celsius}')


````
