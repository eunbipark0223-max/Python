```python
# JSON
url = (f'http://openapi.seoul.go.kr:8088/{api_key}/json/tbCycleRentData/{start}/{end}/2025-01-16/1/') # 데이터 갱신 날짜: 2025-01-16
data = requests.get(url, timeout=10).json() # JSON 형식으로 자동 파싱
# pprint.pprint(data)

data_dict = data.get('rentData', {})
print(f'전체 데이터 건수: {data_dict.get("list_total_count", "알 수 없음")}')

df = pd.DataFrame(data_dict.get('row', []))
df.info()
df
```
