```python
# ì»´í“¨í„°ê°€ ìƒê°í•œ ìˆ«ì ë§ì¶”ê¸°
import random
answer = random.randint(1,100)
print('ì»´í“¨í„°:1ë¶€í„° 100ê¹Œì§€ì˜ ìˆ«ì í•˜ë‚˜ë¥¼ ì •í–ˆìŠµë‹ˆë‹¤')
print('ìˆ«ìë¥¼ ë§ì¶”ì–´ ë³´ì„¸ìš”')

time = 0
while True:
  user_A =int(input('ì •ë‹µì€?:'))
  time += 1
  if user_A == answer:
    break
  elif user_A > answer:
    print(f'ì •ë‹µì€ {user_A}ë³´ë‹¤ ì‘ìŠµë‹ˆë‹¤')
  else:
     print(f'ì •ë‹µì€ {user_A}ë³´ë‹¤ í½ë‹ˆë‹¤')

print(f'ì •ë‹µì…ë‹ˆë‹¤. ì´ë•Œê¹Œì§€ ìœ ì¶”í•œ íšŸìˆ˜:{time}')

```
