### 집합 set
* 집합의 원소는 반드시 '값이 변하지 않는(Hashable)' 상태여야 한다. 리스트나 딕셔너리는 언제든 값이 바뀔 수 있기 때문에 집합의 원소가 될 수 없다
* 중복되지 않은 요소들로 구성된 자료형
* 집합내의 모든 요소들은 순서가 없다
* 인덱싱이나 슬라이싱으로 집합 내의 요소에 접근할수 없다
* set1 = {1,2,3}
* set1 = {} 는 딕셔너리로 인식
* set2 = set()
* 연산자 : | 합집합, & 교집합, - 차집합, ^ 대칭 차집합(합집합에서 교집합을 뺀 집합)
* 내장함수 : add(), remove(), update()


```python
nums = {1,1,1,1,1}
print(nums)
print(len(nums))
```

```python
nums={1,2,3}
print(nums[1]) #집합은 인덱싱, 슬라이싱 요소에 접근할 수 없다
```

```python
set1 = {1,2,3,4,5}
set2={4,5,6}
print(set1|set2) #합집합
print(set1&set2) #교집합
print(set1-set2) #차집합
print(set1^set2) #대칭차집합 (합집합 - 교집합)
print(1 in set1)
print(len(set1))
print({1,2} in set1) #{{1,2},3,4,5}
```

```python
#add():집합에 원소 추가, 이미 있는 경우 추가 안됨
nums = {1,2,3}
nums.add(4)
print(nums)
nums.add(2) #이미 있는 경우 중북 추가 안됨
print(nums)
```

```python
#remove():원소 제거, 집합에 없으면 에러
nums = {1,2,3}
nums.remove(2)
print(nums)
nums.remove(2)
```

```python
#discard() : 원소 제거, 집합에 없어도 에러 발생하지 않음
nums = {1,2,3,4}
nums.discard(4)
print(nums)
nums.discard(5) #에러 발생하지 않음
```

```python
#update():여러개의 원소를 한번에 추가, 입력의 자료형은 리스트,튜플,집합등 사용 가능
nums.update({1,2,4})
print(nums)
```

```python
print(nums[3])
print(nums[:3])
```

### 파이썬의 자료형은 데이터의 값을 생성 후 변결할 수 있는지에 따라 '가변(mutable)' 과 '불변(immtable)'으로 나윔
###불변(immutable)
* int,float,bool,str,tuple
* 만약 값을 바꾸려고 하면, 기존 객체가 변하는것이 아닌, 새로운 값을 가진 객체가 생성되고 변수가 그 새 객체를 가리키게 된다.
* 데이터가 안전하게 보호되며, 딕셔너리의 키로 사용할 수 있다.
* 데이터가 안전하게 보호되며, 딕셔너리의 키로 사용할 수 있다.
값 변경 불가능 , 새 객체 생성
* 값이 바뀌면 메모리 주소도 바뀜
* 상수, 딕셔너리 키에서 사용
* 값 변화가 예측 가능, 안정성이 높음

###가변(mutable)
* list,dict,set
* 객체가 생성된 후에도 내부의 값을 자유롭게 변경 할 수 있는 자료형, 새로운 객체를 만들지 않고 기존 메로리 주소에서 값만 바뀐다.
* 데이터의 추가 , 삭제 , 수정이 빈번한 경우 효율적. 하지만 여러 변수가 동시에 참조할때 예상치 못한 부작용이 발생할 수 있다.
* 값 변경 가능, 기존 객체 수정
* 값이 바뀌어도 메모리 주소는 그대로
* 데이터 모음, 동적 리스트에서 사용
* 참조 공유시 주의 필요, 안정성이 낮



```python
#1 불변 객체
a= 10
b=a
a=a+5
print(a)
print(b)

#2 가변 객체
list_a=[1,2,3]
list_b=list_a
list_a.append(4)
print(list_a)
print(list_b)
```
