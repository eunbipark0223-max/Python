# 📘 파이썬 기초: 01. 프로그래밍과 파이썬



## 1. 프로그래밍 언어
* **프로그래밍의 이해**:
    * **프로그램(Program)**: 컴퓨터로 문제를 해결하기 위해 작성하는 명령어의 모음입니다.
    * **프로그래밍(Programming)**: 프로그램을 작성하는 과정입니다.
    * **프로그래머(Programmer)**: 프로그램을 작성하는 사람을 말합니다.
* **프로그래밍 언어의 종류**:
    * **저급 언어**: 기계 중심의 언어로 하드웨어 지향적입니다. (예: 어셈블리어)
    * **고급 언어**: 사람이 이해하기 쉬운 언어와 기호를 사용합니다. (예: C, 자바, 파이썬 등)
* **언어 번역**: 사람이 작성한 고급 언어를 컴퓨터가 이해할 수 있는 기계어로 변환하는 과정이 필요합니다.

## 2. 파이썬의 개념
* **파이썬(Python)**: 1991년 귀도 반 로섬이 발표한 프로그래밍 언어입니다.
* **주요 특징**:
    * 문법이 간결하고 사람의 언어와 비슷하여 배우기 쉽습니다.
    * 다양한 분야(데이터 분석, 인공지능, 웹 개발 등)에서 활용할 수 있는 풍부한 라이브러리를 제공합니다.
    * 무료로 사용할 수 있는 오픈 소스입니다.

# 파이썬의 특징


*   인기있는 이유: 다양한 라이브러리 활용 가능
*   라이브러리: NumPy(고성능 수치계산),사이킷런(분류,회기,군집화)

#파이썬 변수 이름 규칙
* 첫번째 글자는 문자 또는 밑줄로 시작해야함(숫자가 먼저 올 수 없음)
* 영문자 대소문 구분
* 예약어를 변수로 쓸 수 없음



## 3. 파이썬 개발 환경
* **구글 코랩(Google Colab)**: 웹 브라우저에서 파이썬 코드를 작성하고 실행할 수 있는 서비스입니다.
    * 별도의 설치 없이 무료로 사용할 수 있습니다.
    * 작성한 파일은 구글 드라이브(`Colab Notebooks` 폴더)에 자동으로 저장됩니다.
* **구성 요소**:
    * **코드 셀**: 파이썬 코드를 입력하고 실행하는 공간입니다.
    * **텍스트 셀**: 설명이나 메모를 마크다운 형식으로 작성하는 공간입니다.

## 4. 처음 만나는 프로그램
* **프로그램 실행**: 코드 셀에 소스 코드를 입력한 후 실행 버튼을 누르거나 단축키(`Ctrl + Enter`)를 사용하여 결과를 확인합니다.
* **파일 관리**: 파일 이름을 변경하거나 구글 드라이브를 통해 저장된 파일을 불러올 수 있습니다.

```python
print('파이썬 수업을 시작함')
```

## 5. [실습 예제]



### 💬 문자열 출력
1. **기능**: `print()` 함수를 사용하여 화면에 텍스트를 출력합니다.
2. **로직**: 출력하고 싶은 문장을 따옴표(`' '` 또는 `" "`)로 감싸서 함수 안에 넣습니다.
3. **결과**: 실행 시 따옴표 내부의 글자가 화면에 나타납니다.



```python
print('안녕하세요!')
print("안녕하세요!")
print('안녕' + '파이썬')
print('5' + '<-여기서 5는 문자열')
print('문자열에는'+ '        ' + '띄어쓰기도 넣을 수 있지')
```

### 🔢 연산 결과 출력
1. **기능**: 파이썬을 계산기처럼 사용하여 숫자 연산 결과를 출력합니다.
2. **로직**: `print()` 함수 안에 사칙연산(`+`, `-`, `*`, `/`) 수식을 넣습니다.
3. **결과**: 수식을 계산한 최종 결과값이 화면에 출력됩니다.

```python
print(5*100)
print('5')
```
