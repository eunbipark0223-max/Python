### 클래스

* 클래스의 선언 및 인스턴스 생성
```
class Starbucks:
    brand = "Starbucks"  # 클래스 변수 (모두가 공유)
```

1. 인스턴스 (Instance): 클래스(설계도)를 바탕으로 실제 메모리에 만들어진 실체, '스타벅스 메뉴'라는 설계도로 찍어낸 실제 아메리카노가 인스턴스

2. 메서드 (Method): 클래스 내부에 정의된 함수. 객체가 할 수 있는 '행동'을 정의, 커피 객체의 '주문하기'같은 기능

3. 애트리뷰트 (Attribute): 객체가 가지고 있는 상태나 성질. (인스턴스 변수와 클래스 변수를 모두 포함하는 상위 개념), 커피의 '이름', '가격'같은 정보

4. self: 객체 자기 자신을 가리키는 키워드. 메서드를 호출할 때 어떤 객체가 부른 것인지 구분하기 위해 사용

5. 생성자 (Constructor): 객체가 생성될 때 자동으로 호출되는 특수 메서드(__init__). 주로 초기값을 설정할 때 사용, 커피가 만들어지자마자 컵에 이름표를 붙이고 기본 가격을 매기는 과정

6. 인스턴스 변수 (Instance Variable): 각 인스턴스마다 개별적으로 가지는 변수. self.변수명 형태로 정의

7. 클래스 변수 (Class Variable): 해당 클래스로 만든 모든 인스턴스가 공유하는 변수. 클래스 내부, 메서드 외부에서 정의.

### 매직 메서드 Magic method


```
 __init__() : 생성자의 역할 수행, 인스턴스가 처음 생성될 때부터 필요한 변수 선언 및 메서드 호출을 실행, return 키워드 사용 불가.
 __call__() : 인스턴스를 함수처럼 호출했을 때 실행되는 메서드
 __len__() : len()함수의 입력으로 클래스의 인스턴스를 사용할때 호출되는 함수, self 이외의 인자를 받을 수 없다.
 __getitem__() : 인스턴스에 인덱싱과 슬라이싱을 수행할때 실행되는 함수, self 이외에 하나의 인자를 받고, 대괄호에 사용할 인덱싱, 슬라이싱 정보가 된다.
 __add__() : 덧셈연산자+를 사용하면 자동으로 __add__() 메서드 호출
```



````python
class Number:
  def set_num(self,n):
    self.num = n
  def __add__(self,a):
    return self.num + a

two = Number()
two.set_num(2)
print(two + 5) # 덧셈 연사자+ 를 사용하면 자동으로 __add__() 메서드 호출됨
print(two.__add__(5))# 보통 매직 매서드는 이름을 써서 호출하지 않음
````

````python
class XY:
  def __init__(self,x,y):
    self.x=x
    self.y=y
    print(x,y)

c = XY(3,4)
````

````python
class Multiplier:
  def __init__(self,mul):
    self.mul = mul
  def __call__(self,num):
    return self.mul * num
double = Multiplier(2)
triple = Multiplier(3)

print(double(10))#인스턴스를 함수처럼 사용하여 __call__() 메서드 호출
print(triple(10))
````

````python
class Data:
  def __init__(self,data):
    self.data = [-1]+data
  def __len__(self):
    return len(self.data)

d = Data([10,20,30])
print(len(d)) # __len__() 메서드로 선언, self 이외의 인자를 받을 수 없음
````

````python
class Members:
  def __init__(self,mems):
    self.mems = mems
  def __len__(self):
    return len(self.mems)
  def __getitem__(self,idx):
    return self.mems[idx]

m = Members(['kim','lee','park'])
print(m.mems)
print(len(m))
print(m[0])
````

````python
class FindPath: #부모 클래스
  def find(self,start,dest):
    print(start,dest)

class FPBicycle(FindPath): #자식 클래스
  def get_bicycle(self):
    print('get bicycle road')

bicycle = FPBicycle()
bicycle.find('Seoul','Busan')
````

````python
class FindPath: #부모 클래스
  def find(self,start,dest):
    print(start,dest)

class FPBicycle(FindPath): #자식 클래스
  def find(self,start,dest):
    print('bicycle:' , start,dest)

bicycle = FPBicycle()
bicycle.find('Seoul','Busan')
````

````python
class FindPath: #부모 클래스
  def find(self,start,dest):
    print(start,dest)

class FPBicycle(FindPath): #자식 클래스
  def find(self,start,dest):
    super().find(start,dest) #super(). 부모 클래스 사용
    print('bicycle:' , start,dest)

bicycle = FPBicycle()
bicycle.find('Seoul','Busan')
````

### @property : 클래스의 인스턴스 변수에 접근할때 해당 메서드가 실행된다.


```
class 클래스명:
  @property
  def 변수(self):
    실행코드
  @변수.setter
  def 변수(self, 입력값):
    실행코드
```




````python
class XY:
  def __init__(self,x,y):
    self._x=x # c._x = 3
    self._y=y # c._y = 4
  @property # _x 속성을 그냥 x속성처럼 사용하고 싶을때  (x를 property로 정의 )
  def x(self):
    return self._x

c = XY(3,4)
print(c.x)

````

````python
class XY:
  def __init__(self,x,y):
    self._x=x # c._x = 3
    self._y=y # c._y = 4
  @property # _x 속성을 그냥 x속성처럼 사용하고 싶을때  (x를 property로 정의 )
  def x(self):
    return self._x
  @x.setter #메서드 이름과 동일한 이름에 .setter를 붙인 데코레이터를 작성
  def x(self,value):
    if value < 0:
      print('Warning : value<0')
      self._x=value

c = XY(3,4)
print(c.x)
c.x = -1
print(c.x)
````

````python
class XY:
  def __init__(self,x,y):
    self.__x=x #private 속성
    self.__y=y

c=XY(3,4)
print(c._XY__x) #외부에서 인스턴스를 통해서 __x에 접근할때는 #_클래스이름을 붙이면 .__x에 접근 가능
````

````python
class C:
    def __init__(self, r):
        self._r = r

    @property
    def r(self):
        return self._r
    @r.setter
    def r(self, new_r):
        print('r is changed')
        self._r = new_r


c = C(5)

c.r = 10
print(c.r)

````

### @staticmethod 데코레이터를 사용하면 실행 과정에서 인스턴스 변수 값이나 클래스 변수 값이 변하지 않는 정적 메서드를 선언할 수 있다.


```
class 클래스명:
  @staticmethod
  def 메서드(인자):
    실행코드
```



````python
class XY:
  def __init__(self,x,y):
    self.x=x
    self.y=y
  @staticmethod
  def magnitude(x,y): #self가 없는 정적 메서드는 인스턴스와 클래스에서 각각 호출
    return (x**2+y**2)**(1/2)

c = XY(3,4)
print(c.magnitude(c.x,c.y))
print(XY.magnitude(5,12))
````

### @classmethod 데코레이터는 메서드를 클래스 메서드로 선언하며, 클래스 변수 값을 변경하거나 읽을때 사용


```
class 클래스명:
  @classmethod
  def 메서드(cls, 변수): # 첫번째 인자가 클래스를 가리키며 클래스변수에 접근하는데 사용
    실행 코드
```



````python
class Korean:
  country = 'korea' #클래스 변수
  @classmethod
  def change_country(cls,name): #cls 는 클래스
    cls.country = name

k1 = Korean()
k2 = Korean()
k1.change_country('South Korea')
print(k1.country)
print(k2.country)
````

````python
class Member:
    language = 'python'

    def __init__(self, name):
        self.name = name

    @staticmethod
    def print_upper(name):
        print(name.upper())

    @classmethod
    def change_lang(cls, lang):
        cls.language = lang


m1, m2 = Member('kim'), Member('lee')

m1.print_upper(m1.name)
m1.change_lang('PYTHON')

print(m2.language)
````

```
getattr(객체, 이름[, 기본값])
```

### setattr() 클래스 객체의 애트리뷰트를 동적으로 할당할때 사용하는 파이썬 내장 함수

````python
class XY:
  def __init__(self,x,y):
    self.x=x
    self.y=y
c= XY(3,4)
setattr(c,'z',8) #인스턴스 변수z에 8을 할당
print(c.z)
````

````python
class XY:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __getattr__(self, n):
        print(f'no attribute {n}')

c = XY(3,4)

print(c.z)   # 존재하지 않는 속성
print(c.x)   # 존재하는 속성
````



```

__setattr__(self, name, value): 객체의 속성이 설정될 때 내부적으로 자동 호출되는 매직 메서드. 속성 설정 방식을 커스텀(재정의)하고 싶을 때 사용.
```

```
구분	    setattr() (함수)	                  __setattr__ (매직 메서드)

사용 시점	 속성 이름을 문자열로 가지고 있을 때	      속성이 저장되는 방식을 통제하고 싶을 때
작동 방식	 내부적으로 결국 __setattr__을 실행함	    모든 속성 설정(obj.a = 1) 시 호출됨
```
