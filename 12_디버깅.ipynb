### 디버깅(Debugging) : 프로그램 코드에 있는 오류를 발견하고 수정하는 작업

### 에러 강제 발생
assert 조건식, 에러 메시지
: 조건을 검증할때 사용, 조건식이 False인 경우 AssertionError : 에러 메시지 출력

raise 에러 메시지
: 의도적으로 에러를 발생시킴, 미구현 에러 NotImplementedError

try:
    실행 코드
except:
    예외 처리 코드

try문 : 예외 처리, 프로그램 실행 중 에러가 발생 했을 때 예외처리 코드가 실행된다.

try:
    실행 코드
except exception as e:
    print(e) 에러 정보 출력
    예외 처리 코드

try:
    실행 코드
except:
    예외 처리 코드
else:
    에러가 발생하지 않으면 실행되는 코드
finally:
    에러 발생 여부와 관계없이 실행되는 코드


```python
nums = [1,2,3]
try:
  n = nums[5]
  print(n)
except:
  print("에러 발생")
```

```python
nums = []
for i in range(100):
  try:
    div = i - 50
    nums +=[100/div]
  except Exception as e:
    print(e)
    print(i,div)
```

```python
nums = [ 1,20,300]
try:
  first = nums[0] / 0
except IndexError:
  print('인덱스 범위를 벗어났습니다.')
except ZeroDivisionError:
  print("0으로 나눌 수 없습니다.")
print("프로그램이 정상적으로 종료되었습니다.")
```

```python
string = 'python'
try:
  num = int(string)
except ValueError:
  print('문자열을 정수로 변환할 수 없습니다.')
```

```python
try:
  num = 9 / 3
except:
  print('에러가 발생했습니다.')
else:
  print('에러가 발생하지 않았습니다.')
```

```python
try:
  num = 9/0
except:
  print('에러가 발생했습니다')
else:
  print('에러가 발생하지 않았습니다.')
finally:
  print('에러 발생 여부와 관계없이 실행됩니다.')
```

breakpoint() : 파이썬3.7버전, 프로그램을 멈추고 상호작용하는 디버거를 실행, .py파일에서 실행
(pdb) 명령어 입력
명령어 모음
print(p) 변수의 값을 출력 한다
next(n) 현재 줄의 코드를 실행하여 다음 줄로 넘어간다
continue(c) breakpoint()를 끝내고 나머지 코드 실행
list(l) 현재 실행 되고 있는 파일의 소스 코드를 확인
quit(q) breakpoint()와 프로그램 실행을 모두 종료
