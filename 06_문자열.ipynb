3. **ëª¨ìŠ¤ ë¶€í˜¸ â†’ ë¬¸ì¥**:
    * ì…ë ¥ë°›ì€ ëª¨ìŠ¤ ë¶€í˜¸ë¥¼ ê³µë°± ê¸°ì¤€ìœ¼ë¡œ `split()`í•©ë‹ˆë‹¤.

```python
alphabet = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
morse = ['.-','-...','-.-.','-..','.','..-.','--.',
         '....','..','.---','-.-','.-..','--','-.',
         '---','.--.','--.-','.-.','...','-','..-',
         '...-','.--','-..-','-.--','--..']

# ğŸ”¹ ì•ŒíŒŒë²³ â†’ ëª¨ìŠ¤ë¶€í˜¸
s = input('ì•ŒíŒŒë²³ ë¬¸ìì—´: ').upper()   # ì†Œë¬¸ì ëŒ€ë¹„

for ch in s:
    if ch in alphabet:
        print(morse[alphabet.index(ch)], end=' ')
print()

# ğŸ”¹ ëª¨ìŠ¤ë¶€í˜¸ â†’ ì•ŒíŒŒë²³
m = input('ëª¨ìŠ¤ ë¶€í˜¸ ì…ë ¥ (ê³µë°±ìœ¼ë¡œ êµ¬ë¶„): ')
m_list = m.split()

for code in m_list:
    if code in morse:
        print(alphabet[morse.index(code)], end='')
```
